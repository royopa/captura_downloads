```python
import pandas as pd
from datetime import datetime, timedelta

to_date_str = lambda x: (datetime(1900, 1, 1) + timedelta(x - 2)).strftime("%Y-%m-%d")
```

```python
df = pd.read_excel(
    "DADOS DIÁRIOS - ANBIMA - 2021_07_15.xlsb",
    engine="pyxlsb",
    sheet_name="ANBIMA - DEBÊNTURES",
)
```

```python
df
```

```python
names = [
    "symbol",
    "name",
    "maturity_date",
    "underlying",
    "bid_yield",
    "ask_yield",
    "ref_yield",
    "price",
    "perc_price_par",
    "duration",
    "perc_reune",
    "ref_ntnb",
    "refdate",
]
```

```python
# Index(['DATA BASE', 'Código', 'Nome', 'Repac./  Venc.', 'Índice/ Correção',
#        'Taxa de Compra', 'Taxa de Venda', 'Taxa Indicativa', 'Desvio Padrão',
#        'Min.', 'Máx.', 'PU', '% PU Par', 'Duration', '% Reune',
#        'Referência NTN-B'],
#       dtype='object')

df1 = df.iloc[:, [1, 2, 3, 4, 5, 6, 7, 11, 12, 13, 14, 15, 0]].copy()
df1
```

```python
df1.columns = names
```

```python
df1
```

```python
df2 = df1.copy()
df2["ask_yield"] = pd.to_numeric(df1["ask_yield"], errors="coerce").astype("float64")
df2["bid_yield"] = pd.to_numeric(df1["bid_yield"], errors="coerce").astype("float64")
df2["ref_yield"] = pd.to_numeric(df1["ref_yield"], errors="coerce").astype("float64")
df2["price"] = pd.to_numeric(df1["price"], errors="coerce").astype("float64")
df2["perc_price_par"] = pd.to_numeric(df1["perc_price_par"], errors="coerce").astype(
    "float64"
)
df2["duration"] = pd.to_numeric(df1["duration"], errors="coerce").astype("float64")
df2["perc_reune"] = pd.to_numeric(df1["perc_reune"], errors="coerce").astype("float64")
df2["refdate"] = df1["refdate"].map(to_date_str)
df2.loc[~df1["ref_ntnb"].isna(), "ref_ntnb"] = df1.loc[
    ~df1["ref_ntnb"].isna(), "ref_ntnb"
].map(to_date_str)
df2["maturity_date"] = df1["maturity_date"].map(to_date_str)

x = ~df2["ref_yield"].isna()
df2 = df2[x].copy()

df2
```
