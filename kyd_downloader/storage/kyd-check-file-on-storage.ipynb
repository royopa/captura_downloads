```python
import tempfile
import os

import pandas as pd
import google.cloud.storage as storage
from google.cloud.storage.blob import Blob

storage_client = storage.Client()
```

```python
filename = 'gs://ks-layer1/B3StockIndexInfo/2023-05-31.parquet'
blob = Blob.from_string(filename, client=storage_client)

handle, foutput = tempfile.mkstemp()
tempf = os.fdopen(handle, 'w+b')
storage_client.download_blob_to_file(blob, tempf)
tempf.flush()
tempf.seek(0)
df = pd.read_parquet(tempf)
df
```
